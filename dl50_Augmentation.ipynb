```python
from keras.preprocessing import image #behaves like opencv, helps in working on images
from keras.preprocessing.image import ImageDataGenerator
```

```python
img=image.load_img('/content/cat.jpg', target_size=(200, 200))
```

```python
datagen=ImageDataGenerator(
    rotation_range=30, # by how much angle to rotate
    shear_range=0.2, # 1 k scale me 0.2 ka distort
    zoom_range=0.2, # max limit of zoom-in/out
    horizontal_flip=True,
    # vertical_flip=True, do not flip cuz cats don't look like feet upwards
    width_shift_range=0.2, # l-r khiskaya
    height_shift_range=0.2, #up-down khiskaya i.e. shift kiya
    fill_mode='nearest' # jab shift hoega image, nearest wale pixels a aa jenge to fill !! ='reflect' jab shift hoega image, reflection aa jenge to fill !! ='constant' hate hue ki jagah black pixels aa jaenge
    )
```

```python
img=image.img_to_array(img)
```

```python
input_batch=img.reshape(1, 200, 200, 3) # 1 imge of 200,200,3
```

```python
#image liya, numpy array me transform kiya, shape change kiya
# bject create kiya to tell kya transformation apply krne he
```

```python
i=0
#flow is used for a single image, flow_for_directory when folder
# batch_size=1 i.e o/p me 1 batch or form hoega
# save_to_dir- name where to save
for output in datagen.flow(input_batch, batch_size=1, save_to_dir='/var/tmp'):
  i+=1
  if i==10:
    break
```
